# Running Assembly Using FLYE

```python
![ -e 'Out Canu' ] && rm -r 'Out Canu'
!mkdir 'Out Canu'

!/media/anuradhawick/data/Tools/canu/Linux-amd64/bin/canu \
-p all \
-pacbio-raw 'Sampled Reads/sd_0001.fastq' 'Sampled Reads/sd_0002.fastq' 'Sampled Reads/sd_0003.fastq' 'Sampled Reads/sd_0004.fastq' \
-d 'Out Canu' \
genomeSize=6m cnsMemory=10 cnsConcurrency=5 \
```

# Generating Assembly Graph

```python
!cd 'Out Canu' && /media/anuradhawick/data/Tools/Bandage/Bandage_Ubuntu_dynamic_v0_8_1/Bandage \
image all.unitigs.gfa ../image_graph_canu_all.png --names --lengths --depth --height 1080 --fontsize 3
```
