# Generating Two Haplotypes

```python
import random

from Bio.Seq import Seq
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import DNAAlphabet

heterozygous_rate = 0.01

alpha = set(['A', 'C', 'G', 'T'])
records = []
for seq_record in SeqIO.parse("/media/anuradhawick/data/Experiments/Assembly_Graph/Assembly/Sample_Genomes/yeast/all_chr.fa", "fasta"):
    s_len = len(seq_record)
    s = seq_record.seq
    sl = list(s)
    delta = round(s_len * heterozygous_rate)
    edit_pos = random.sample(range(0, s_len), delta)
    
    for edit in edit_pos:
        o = sl[edit]
        new_alpha = alpha - set([o])
        new_char = list(new_alpha)[random.randint(0,2)]
        sl[edit] = new_char
    
    record = SeqRecord(Seq("".join(sl), DNAAlphabet), id=seq_record.id + "_h2", description="other hap")    
    records.append(record)

SeqIO.write(records, "/media/anuradhawick/data/Experiments/Assembly_Graph/Assembly/Sample_Genomes/yeast/hap2.fa", "fasta")
    
!cat \
/media/anuradhawick/data/Experiments/Assembly_Graph/Assembly/Sample_Genomes/yeast/all_chr.fa \
/media/anuradhawick/data/Experiments/Assembly_Graph/Assembly/Sample_Genomes/yeast/hap2.fa \
> /media/anuradhawick/data/Experiments/Assembly_Graph/Assembly/Sample_Genomes/yeast/with_hap.fasta
    
    
    
    
    
```
